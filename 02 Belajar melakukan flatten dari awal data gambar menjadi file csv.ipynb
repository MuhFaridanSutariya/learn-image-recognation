## mendapatkan semua image path

```python
from glob import glob
```

```python
# glob mencapture pattern dari semua pathnya belum menarik imagenya
fname = glob("data/*/*.png")
fname[:10]
```

## X = Path -> Image -> Flatten

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([plt.imread(f).flatten() for f in fname])
X
```

```python
X.shape
```

## y = Path -> Split -> ambil [1]

```python
y = np.array([int(f.split("\\")[1]) for f in fname])
y
```

```python
#### Kenapa terdapat list 1?. Karena kita ingin mengambil posisi dibagian namanya saja
'data\\0\\1.png'.split("\\")[1]
```

```python
y.shape
```

## Join sebagai 1 dataframe lalu save as csv

```python
# X
df = pd.DataFrame(X, columns=range(1, 785))
df.head()
```

```python
# y
df['label'] = y
df.head()
```

## save to csv

```python
df.to_csv("latihan_metadata.csv", index=None)
```
